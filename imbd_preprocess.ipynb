```python
# librerías 
import re
import os
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# Cargamos las películas
imbd_movies_df = pd.read_csv("./imbd_movies.tsv", sep="\t")
print(imbd_movies_df.info())
display(imbd_movies_df.tail())
```

```python
imbd_movies_df['titleType'].unique()
```

```python
exclu_type = ['video', 'videoGame', 'tvEpisode']
# Nos quedamos con los registros que no son 
imbd_movies_df = imbd_movies_df[~imbd_movies_df['titleType'].isin(exclu_type)]
imbd_movies_df['titleType'].unique()
```

```python
imbd_movies_df.info()
```

```python
# Cargamos los ratings de las películas
imbd_rating_df = pd.read_csv("./imbd_rating.tsv", sep="\t")
print(imbd_rating_df.info())
display(imbd_rating_df.tail())
```

```python
# Unimos los datos de imbd de películas y rating 
# a través de tconst que es un código único.
imbd_final_df = pd.merge(imbd_movies_df, imbd_rating_df, on="tconst")
print(imbd_final_df.shape)
display(imbd_final_df.head())

# Guardamos en un CSV los datos de películas y rating
imbd_final_df.to_csv("./imdb_dataset.csv", sep=";", index=False)
```
